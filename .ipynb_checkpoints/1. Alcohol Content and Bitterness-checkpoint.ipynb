# Alcohol Content and Bitterness

What is the relationship between the alcohol content of a beer and its bitterness? In this part of the lab, you will investigate this relationship.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_rows', 15)
%matplotlib inline

data = pd.read_csv("/data/beer_train.csv")
```

```python
scores = []
```

## Question 1

Find the optimal polynomial model for predicting IBU from ABV. Make a scatterplot of these two variables, with your fitted polynomial curve overlaid.

_Hint:_ You may want to remove certain outliers from the data before fitting your model.

```python
# YOUR CODE HERE
data = data[data['ibu'] < 150]
```

```python
model = LinearRegression()
abv = ['abv']
for i in range(2,12):
    print('Power: %s' %str(i-1))
    print(-cross_val_score(model,data[abv], data['ibu'], cv = 10,
          scoring = "neg_mean_squared_error").mean())
    new_col = "abv^%s" % str(i)
    data[new_col] = data['abv'] ** i
    abv.append(new_col)
```

```python
#The optimal polynomial model is the seventh degree polynomial model
abv = abv[:7]
abv

model.fit(data[abv], data['ibu'])

x_new = np.arange(0,20, 0.2)
X_new = pd.DataFrame({
        "abv": x_new,
    })

for i in range(2,8):
    new = 'abv^%s' % str(i)
    X_new[new] = x_new ** i
```

```python
y_new = model.predict(X_new)
data.plot.scatter(x='abv', y='ibu')
plt.plot(x_new, y_new, 'r-')
```
